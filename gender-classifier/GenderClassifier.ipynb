```python
import pandas as pd  
import numpy as np  
from sklearn.feature_extraction.text import CountVectorizer
import os
from nltk import word_tokenize  
from nltk.util import ngrams
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import nltk
nltk.download('punkt')
```

```python
def data_preparation(filename,category):
    with open(filename , 'r') as f:
        first_line = f.readline()            
        data = []
        for line in f: 
            data.append([line.split("\t")[1].strip().replace(",",""), category] )
    return data

male = data_preparation('./male50K.tsv' ,"M")
female = data_preparation('./female50K.tsv', "F")
data =  male + female
```

```python
all = pd.DataFrame.from_records(data, columns=['name', 'gender'])
print(all.columns)
print(all.size)
print(all.shape)
print(all.dtypes)
print(all.isnull().isnull().sum())
print(all.gender.unique())
print(all.name.unique())
    
  
cv = CountVectorizer( )
X = cv.fit_transform(all['name'])
cv.get_feature_names()
```

```python
for i in list(ngrams("tokens", 3)):
    print(''.join(i)) 
```

```python
from sklearn.model_selection import train_test_split
#Features
X
#label
y = all["gender"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
from sklearn.naive_bayes import MultinomialNB
clf= MultinomialNB()
clf.fit(X_train,y_train)
print("Acc: " , clf.score(X_test,y_test) * 100, "%")
```

```python
print("Acc: " , clf.score(X_train,y_train) * 100, "%")
```

```python
sample = ["Mary","John","Samur","Maria","Peter","Artur","Nikita","Natasha"]
vect = cv.transform(sample).toarray()
clf.predict(vect)
```

```python
def gender_predictor(name):
        test_name=[name]
        vector = cv.transform(test_name).toarray()
        if clf.predict(vector) == "M":
            print("Male")
        else:
            print("Female")
```

```python
gender_predictor("maria da silva")
```
